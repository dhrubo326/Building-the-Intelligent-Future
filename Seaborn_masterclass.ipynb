# Numerical plots

##1.   Distribution plot ( one feature)
##2.   Joinplot (two feature)
##3.   Pairplot (more than two feature)


```python
#distribution plot
sns.histplot(data=df, x='bill_length_mm', kde=True, hue='sex')
#sns.histplot(x='island', data=df, kde=True)
#sns.kdeplot(x='bill_length_mm', data=df)
```

```python
# jointplot
sns.jointplot(data=df, x="bill_length_mm", y="bill_depth_mm", hue="species")
```

```python
#pairplot
# df = sns.load_dataset('tips')
sns.pairplot(data=df, kind='reg')
```

#Categorical data
##1. Countplot
##2. Barplot
##3. Boxplot
##4. Violinplot

```python
sns.countplot(data=df, x='island', palette='PuBu')
```

```python
# both categorical and numerical data needed
sns.barplot(df, x="flipper_length_mm", y="sex",  hue='island', orient='hx')
```

```python
#boxplot
sns.boxplot(y="flipper_length_mm", x='island', hue='species', data=df)
```

```python
#violineplot
sns.violinplot(data=df, x="flipper_length_mm", y="sex", legend='auto',  split=True)
```

#Pairgrid

```python
#pairGrid
g = sns.PairGrid(df)
# g.map(sns.scatterplot)
g.map_diag(sns.histplot)
g.map_offdiag(sns.kdeplot)
g.add_legend()
```
