```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = [10, 20, 30, 40, 50, 60]
y = [100, 200, 300, 400, 550, 260]
```

```python
plt.xlabel("Price")
plt.ylabel("amount")
plt.title('This a matplotlib chart')
plt.plot(x, y, 'b-.o')
plt.show()
```

```python
x1 = [10, 20, 30, 40, 50, 60]
x2 = [5, 55, 77, 99, 111, 133]
y1 = [100, 200, 300, 400, 500, 600]
y2 = [50, 60, 70, 80, 90, 100]
y3 = [200, 400, 600, 800, 1000, 1200]
```

```python

plt.xlabel("Price")
plt.ylabel("amount")
plt.title('Matplotlib chart')

plt.plot(x1, y1, color='r', label='Apple')
plt.plot(x1, y2, color='blue', label='Mango')
plt.plot(x1, y3, '^g--', label='Banana')
# #plt.plot(x2, y2, color='black')
plt.legend(loc='best', shadow=True, fontsize='large')
plt.grid()
plt.show()
```

```python
x1 = [10, 20, 30, 40, 50, 60]
x2 = [5, 55, 77, 99, 111, 133]
y1 = [100, 200, 300, 400, 500, 600]
y2 = [50, 60, 70, 80, 90, 100]
y3 = [200, 400, 600, 800, 1000, 1200]
plt.xlabel("Price")
plt.ylabel("amount")
plt.title('Matplotlib chart')

plt.subplot(2,2,1)
plt.plot(x1, y1, color='red', label='Apple')
plt.grid()

plt.subplot(2,2,2)
plt.plot(x1, y2, color='blue', label='Mango')

plt.subplot(2,2,4)
plt.plot(x1, y3, '*g--', label='Banana')
plt.grid()

#plt.plot(x2, y2, color='black')
# plt.legend(loc='best', shadow=True, fontsize=11)
plt.show()
```
